Mask the card number by masking first 3 segments: For example: XXXX-XXXX-XXXX-1234

Qno 2. 
We found an issue in the production system.
The log is showing various errors and error codes.
The format of log is consistent and the pattern is always looks like this:
 
INFO 200: User login successful at 09:15AM
INFO 201: New user registered at 09:20AM
WARNING 301: Resource moved temporarily at 09:45AM
WARNING 408: Request timeout at 10:05AM
ERROR 404: Page not found at 10:45PM
ERROR 500: Internal server error at 11:10PM
 
From this log, extract only the errors and the error codes at error timings
 
How many logs are there for each: info, warning, error

```python
logs = (
    "INFO 200: User login successful at 09:15AM",
    "INFO 201: New user registered at 09:20AM",
    "WARNING 301: Resource moved temporarily at 09:45AM",
    "WARNING 408: Request timeout at 10:05AM",
    "ERROR 404: Page not found at 10:45PM",
    "ERROR 500: Internal server error at 11:10PM"
)
info_count = 0
warning_count = 0
error_count = 0
print("Error Logs:")
for log in logs:
    if log.split()[0] == "INFO":
        info_count += 1

    elif log.split()[0] == "WARNING":
        warning_count += 1

    elif log.split()[0] == "ERROR":
        error_count += 1
        print(
            "Error Code:", log.split()[1].replace(":", ""),
            "Time:", log.split()[-1]
        )

print("\nSummary:")
print("INFO logs:", info_count)
print("WARNING logs:", warning_count)
print("ERROR logs:", error_count)
```

```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
.re
# Difference: elements in set1 but not in set2
set1.difference(set2)
# Output: {1, 2}

# Intersection: common elements
set1.intersection(set2)
# Output: {3, 4}

# Union: all unique elements from both sets
set1.union(set2)
# Output: {1, 2, 3, 4, 5, 6}

# Subset: is every element of set1 in set2?
set1.issubset(set2)
# Output: False

# Superset: does set1 contain all elements of set2?
set1.issuperset(set2)
# Output: False
```

## Operators

### Arithmatic

| Operator | Description        | Example        |
|---------:|--------------------|----------------|
| `+`      | Addition           | `5 + 3 = 8`    |
| `-`      | Subtraction        | `5 - 3 = 2`    |
| `*`      | Multiplication     | `5 * 3 = 15`   |
| `/`      | Division           | `5 / 3 = 1.66` |
| `//`     | Floor Division     | `5 // 3 = 1`   |
| `%`      | Modulus            | `5 % 3 = 2`    |
| `**`     | Exponentiation     | `5 ** 3 = 125` |

### Comparison Operator

| Operator | Description                    | Example            |
|:--------:|--------------------------------|--------------------|
| `==`     | Equal to                       | `5 == 3 → False`   |
| `!=`     | Not equal to                   | `5 != 3 → True`    |
| `>`      | Greater than                   | `5 > 3 → True`     |
| `<`      | Less than                      | `5 < 3 → False`    |
| `>=`     | Greater than or equal to       | `5 >= 3 → True`    |
| `<=`     | Less than or equal to          | `5 <= 3 → False`   |


### Logical Operator

| Operator | Description                                      | Example                  |
|:--------:|--------------------------------------------------|--------------------------|
| `and`    | Returns True if both expressions are True        | `True and False → False` |
| `or`     | Returns True if one of the expressions is True   | `True or False → True`   |
| `not`    | Reverses the boolean value                       | `not True → False`       |
